```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI(
    model='gpt-5.4-mini'
)
```

```python
class PromptChaining(TypedDict):
    topic: str
    topic_outline: str
    outline_blog: str

```

```python
# outline
def llm_outline(state: PromptChaining) -> PromptChaining:

    topic = state['topic']
    response = model.invoke(
            f"Generate a detailed outline for {topic}"
        )    
    state['topic_outline'] = response.content

    return state
    
```

```python
def llm_blog(state: PromptChaining) -> PromptChaining:
    topic_outline = state['topic_outline']

    response = model.invoke(
        f"Generate me a Blog for the {topic_outline}"
    )
    state['outline_blog'] = response.content

    return state

    
```

```python
graph = StateGraph(PromptChaining)

# add node
graph.add_node('llm_outline', llm_outline)
graph.add_node('llm_blog', llm_blog)

# add edges 
graph.add_edge(START, 'llm_outline')
graph.add_edge('llm_outline', 'llm_blog')
graph.add_edge('llm_blog', END)

# compile
workflow = graph.compile()
```

```python
result = workflow.invoke({"topic":"Narendra Modi"})
```
